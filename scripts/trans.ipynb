```python
import torch
from transformers import AutoProcessor, AutoModelForVision2Seq
from PIL import Image
import time


MODEL = "/home/hasan4/Models/openvla"

processor = AutoProcessor.from_pretrained(MODEL, trust_remote_code=True)
vla = AutoModelForVision2Seq.from_pretrained(
    MODEL,
    torch_dtype=torch.bfloat16,
    low_cpu_mem_usage=True,
    trust_remote_code=True
).to("cuda" if torch.cuda.is_available() else "cpu")

# Helper function to load image
def load_image(image_path: str) -> Image.Image:
    return Image.open(image_path)

# Load image from file
image_path = "image.jpg"  # Change this to your image file path
image = load_image(image_path)

# Format the prompt
instruction = "pick up the object"  # Example instruction
prompt = f"In: What action should the robot take to {instruction}?\nOut:"

# Number of inferences to run
num_inferences = 10  # Adjust this as needed

# Start timing the multiple inferences
total_time = 0
total_tokens = 0

for _ in range(num_inferences):
    start_time = time.time()

    # Process the image and instruction
    inputs = processor(prompt, image).to("cuda" if torch.cuda.is_available() else "cpu", dtype=torch.bfloat16)

    # Run the model to predict the action
    action = vla.predict_action(**inputs, unnorm_key="bridge_orig", do_sample=False)

    # End timing the inference
    end_time = time.time()

    # Calculate time taken for inference
    inference_time = end_time - start_time
    total_time += inference_time

    # Calculate tokens generated for the inference
    num_tokens = action.shape[-1]  # The last dimension should give us the number of tokens
    total_tokens += num_tokens

# Calculate average time and tokens per second (TPS)
average_inference_time = total_time / num_inferences
tokens_per_second = total_tokens / total_time if total_time > 0 else 0

# Print the results
print(f"Average Inference Time: {average_inference_time:.4f} seconds")
print(f"Tokens per Second (TPS): {tokens_per_second:.2f} tokens/s")

# Optionally, print the predicted action of the last inference
print(f"Predicted Action (last run): {action}")
```
